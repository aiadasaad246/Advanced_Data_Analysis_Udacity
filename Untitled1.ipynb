```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Target website
url = "http://quotes.toscrape.com"

# Send GET request
response = requests.get(url)



response.text
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
soup
```

```python
# Find all quote containers
quotes = soup.find_all("div", class_="quote")

quotes
```

```python
quote_data = []

for quote in quotes:
    # Extract quote text
    text = quote.find("span", class_="text").text
    print(text)
```

```python
for quote in quotes:
    author = quote.find("small", class_="author").text#adding .text
    print(author)
```

```python
for quote in quotes:
    print(quote.find_all("a", class_="tag"))
    tags = [tag.text for tag in quote.find_all("a", class_="tag")]
    print(tags)
```

```python
quote_data=[]
for quote in quotes:
    # Extract quote text
    text = quote.find("span", class_="text").text

    # Extract author
    author = quote.find("small", class_="author").text

    # Extract tags
    tags = [tag.text for tag in quote.find_all("a", class_="tag")]

    # Append to list
    quote_data.append({"Quote": text, "Author": author, "Tags": ", ".join(tags)})

# Convert to DataFrame and save
df = pd.DataFrame(quote_data)
df.head()
```
